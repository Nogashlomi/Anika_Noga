```python
import pandas as pd 
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_to_xlsx = 'G:\\My Drive\\DataFiles\\Full_Image_List_6.5.24\\all_elements.xlsx'
```

```python
df = pd.read_excel(path_to_xlsx)
```

```python
df
```

```python
# Group by cluster_name and count unique images
unique_images_per_cluster = df.groupby('cluster_name')['images'].nunique()

# Define the bins for the ranges
bins = [0, 1, 5, 10, 20, 50, float('inf')]
labels = ['1 image', '2-5 images', '6-10 images', '11-20 images', '21-50 images', '51+ images']

# Cut the counts into the defined bins
cluster_ranges = pd.cut(unique_images_per_cluster, bins=bins, labels=labels, right=True)

# Count the number of clusters in each range
range_counts = cluster_ranges.value_counts().sort_index()

# Plotting
range_counts.plot(kind='bar')
plt.xlabel('Number of Unique Images in Cluster')
plt.ylabel('Number of Clusters')
plt.title('Distribution of Clusters by Number of Unique Images')
plt.show()
```

```python
# Group by cluster_name and count unique images
unique_images_per_cluster = df.groupby('cluster_name')['images'].nunique()

# Define the bins for the ranges
bins = [0, 1, 5, 10, 20, 50, float('inf')]
labels = ['1 image', '2-5 images', '6-10 images', '11-20 images', '21-50 images', '51+ images']

# Cut the counts into the defined bins
cluster_ranges = pd.cut(unique_images_per_cluster, bins=bins, labels=labels, right=True)

# Count the number of clusters in each range
range_counts = cluster_ranges.value_counts().sort_index()

# Plotting a pie chart
plt.figure(figsize=(10, 6))
plt.pie(range_counts, labels=range_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Clusters by Number of Unique Images')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

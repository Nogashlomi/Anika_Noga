```python
#how many images were printed in total?
# Count unique values in 'column_name'
unique_values = df['images'].unique()
unique_count = len(unique_values)

print(unique_count)
```

```python
#number of images printed by 20 year intervals

# Step 1: Group and count unique images by 20-year intervals
intervals = pd.cut(df['year'], bins=range(1470, 1650, 20), right=False)
interval_count_images = df.groupby(intervals, observed=False)['images'].nunique()

# Step 2: Plotting
plt.figure(figsize=(10, 6))
plt.bar(interval_count_images.index.astype(str), interval_count_images.values, width=0.8, align='center')
plt.title('Number of Printed Images')
plt.xlabel('20-Year Intervals')
plt.ylabel('Number of Unique Images')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
#how many books were printed in total?

# Count unique values in 'book'
unique_book_values = df['book'].unique()
unique_book_count = len(unique_book_values)

print(unique_book_count)
```

```python
#number of book pritned in 20 year intervals
#these are only books that have images!! there are 17 books without any images.

# Step 1: Group and count unique book by 20-year intervals
intervals = pd.cut(df['year'], bins=range(1470, 1650, 20), right=False)
interval_count_books = df.groupby(intervals, observed=False)['book'].nunique()

# Step 2: Plotting
plt.figure(figsize=(10, 6))
plt.bar(interval_count_books.index.astype(str), interval_count_books.values, width=0.8, align='center')
plt.title('Number of Printed Books (books with images)')
plt.xlabel('20-Year Intervals')
plt.ylabel('Number of Unique Images')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
#image per book over the years

# Define the 20-year intervals based on 'year'
df['year_interval'] = pd.cut(df['year'], bins=range(1470, 1650, 20), right=False)

# Calculate number of unique images and books per interval
interval_counts = df.groupby('year_interval').agg({
    'images': 'nunique',
    'book': 'nunique'
}).rename(columns={'images': 'unique_images', 'book': 'unique_books'})

# Calculate the ratio of unique images to unique books
interval_counts['image_to_book_ratio'] = interval_counts['unique_images'] / interval_counts['unique_books']

# Plotting the ratio
plt.figure(figsize=(10, 6))
plt.plot(interval_counts.index.astype(str), interval_counts['image_to_book_ratio'], marker='o', linestyle='-', color='b')
plt.title('Number of Images per Book')
plt.xlabel('20-Year Intervals')
plt.ylabel('Image to Book Ratio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#calculating the number of images per book in a different way
# Group by 'book' and aggregate
book_summary = df.groupby('book').agg(
    unique_images_count=('images', 'nunique'),
    printed_year=('year', 'first')  # Assuming each book has a single year value, otherwise use a suitable aggregation method
).reset_index()
```

```python
book_summary
```

```python
# Create a new column for 20-year intervals
book_summary['year_interval'] = pd.cut(book_summary['printed_year'], bins=range(1470, 1650, 20), right=False)

# Group by the 20-year intervals and calculate the sum of unique images and unique books
interval_summary = book_summary.groupby('year_interval').agg(
    total_unique_images=('unique_images_count', 'sum'),
    total_unique_books=('book', 'nunique')
).reset_index()

# Calculate the ratio of unique images to unique books
interval_summary['image_to_book_ratio'] = interval_summary['total_unique_images'] / interval_summary['total_unique_books']

# Plotting the results
plt.figure(figsize=(12, 8))
plt.plot(interval_summary['year_interval'].astype(str), interval_summary['image_to_book_ratio'], marker='o', linestyle='-', color='b')
plt.title('Ratio of Unique Images to Unique Books Printed in 20-Year Intervals')
plt.xlabel('20-Year Intervals')
plt.ylabel('Ratio of Unique Images to Unique Books')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# compare the two colums to see why there are more links than titles

# Group by 'part_or_adaption_label' and aggregate unique 'part_or_adaption' values
grouped = df.groupby('part_or_adaption_label')['part_or_adaption'].unique()

# Find 'part_or_adaption_label' with multiple matching 'part_or_adaption' values
multiple_matches = {}
for label, values in grouped.items():
    if len(values) > 1:
        multiple_matches[label] = values

# Print the results
if multiple_matches:
    print("part_or_adaption_label with multiple matching part_or_adaption values:")
    for label, values in multiple_matches.items():
        print(f"part_or_adaption_label: {label}, Corresponding Values: {', '.join(values)}")
else:
    print("No part_or_adaption_label with multiple matching part_or_adaption values found.")
```

```python
#create a line bar: one for total number of images printed, the second for the number of images printed in the Tractatus parts in that time frame 

# Function to calculate 20-year intervals
def calculate_20_year_intervals(start_year, end_year):
    intervals = range(start_year, end_year, 20)
    return [(start, start + 19) for start in intervals]

# Define the intervals from 1470 to 1650
intervals = calculate_20_year_intervals(1470, 1650)

# Function to count images in each 20-year interval
def count_images_per_interval(df, intervals):
    results = []
    for interval in intervals:
        start, end = interval
        interval_df = df[(df['year'] >= start) & (df['year'] <= end)]
        total_images = interval_df['images'].nunique()
        tractatus_images = interval_df[interval_df['part_or_adaption_label'] == 'Tractatus de sphaera']['images'].nunique()
        results.append((start, end, total_images, tractatus_images))
    return pd.DataFrame(results, columns=['Start Year', 'End Year', 'Total Images', 'Tractatus de sphaera Images'])

# Calculate images per interval
df_result = count_images_per_interval(df, intervals)
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Plot total images
plt.plot(df_result['Start Year'], df_result['Total Images'], marker='o', linestyle='-', color='b', label='Total Images')

# Plot Tractatus de sphaera images
plt.plot(df_result['Start Year'], df_result['Tractatus de sphaera Images'], marker='s', linestyle='-', color='g', label='Tractatus de sphaera Images')

# Add labels and title
plt.xlabel('Year Intervals')
plt.ylabel('Number of Images')
plt.title('Total and Tractatus de sphaera Images over 20-Year Intervals')
plt.xticks(df_result['Start Year'])
plt.grid(True)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

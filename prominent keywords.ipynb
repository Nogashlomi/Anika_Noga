```python
import pandas as pd 
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_to_xlsx = 'D:\\17.5.24\\DataFiles\\Full_Image_List_6.5.24\\all_elements.xlsx'
```

```python
df = pd.read_excel(path_to_xlsx)
```

```python
#importing the full list of elements and their keywords, years, cluster name, and other info exported

df
```

```python
#making a list of keywords and their related data. the new df name is "grouped_keywords"

# Step 1: Split keywords by semicolon and explode rows
df['keywords'] = df['keywords'].str.split(';')
df_exploded = df.explode('keywords').reset_index(drop=True)

# Step 2: Group by 'keywords' and aggregate 'year' data into a list
grouped_keywords = df_exploded.groupby('keywords').agg(
    total_images=('images', 'size'),  # Count the number of images per keyword
    years_list=('year', lambda x: sorted(x.unique().tolist()))  # Get a sorted list of unique years per keyword
).reset_index()
```

```python
grouped_keywords
```

```python
#add a keyword for "structure of the cosmos" as we don't have UCKs here, and we will need it later

# Keywords to sum
keywords_to_sum = ['CK_07 Spheres', 'CK_08 Spheres', 'CK_09 Spheres', 'CK_10 Spheres', 'CK_11 Spheres', 'CK_12 Spheres']

# Filter the DataFrame to get the relevant rows
filtered_df_for_cosmos_structure = grouped_keywords[grouped_keywords['keywords'].isin(keywords_to_sum)]

# Compute the sum of total_images
total_images_sum = filtered_df_for_cosmos_structure['total_images'].sum()

# Aggregate the years_list
years_list_agg = sorted(set(sum(filtered_df_for_cosmos_structure['years_list'], [])))  # Flatten and get unique years

# Create a new row
new_keyword = 'CK_Structure of the Cosmos'
new_row = {'keywords': new_keyword, 'total_images': total_images_sum, 'years_list': years_list_agg}

# Create a new row
new_keyword = 'CK_Structure of the Cosmos'
new_row = pd.DataFrame({'keywords': [new_keyword], 'total_images': [total_images_sum], 'years_list': [years_list_agg]})

# Append the new row to the DataFrame using pd.concat
grouped_keywords = pd.concat([grouped_keywords, new_row], ignore_index=True)
```

```python
grouped_keywords
```

```python
#"filtered_df" is a df with only CK keywords

filtered_df = grouped_keywords[
    (grouped_keywords['keywords'].str.startswith('CK'))
]
```

```python
filtered_df
```

```python
#ploting the "filtered df" in a pie chart by size

# Plotting
plt.figure(figsize=(8, 6))

# Extracting data for plotting
keywords = filtered_df['keywords']
total_images = filtered_df['total_images']

# Creating pie chart
plt.pie(total_images, labels=keywords, autopct='%1.1f%%', startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Total Images by Keywords')

plt.show()
```

```python
# Find largest groups based on total_images
largest_groups = filtered_df.nlargest(80, 'total_images')  # Get top 3 largest groups, adjust '3' as needed

print("Largest Groups based on Total Images:")
print(largest_groups)
```
